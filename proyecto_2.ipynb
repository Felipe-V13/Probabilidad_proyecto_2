### Probabilidad del n√∫mero de veces que salga la cara de corona en las 10 tiradas del experimento

```python
import matplotlib.pyplot as plt
import numpy as np

def factorial(n):
    fact = 1
    for num in range(2, n + 1):
        fact *= num
    return fact

def binomial_fmp(n, p):
    X = []

    for x in range(n + 1):
        Comb_n_x = factorial(n)/(factorial(x) * factorial(n - x))
        X.append(Comb_n_x * (p**x) * ( (1-p)**(n-x) ))

    return X

n = 10
p = 0.5

print("\nFMP para moneda ideal (p_corona = 0.5): \n")
X_ideal = binomial_fmp(n, p)
for x in range(len(X_ideal)):
    print("P( X =", x, ") = f(", x, ") = ", X_ideal[x])

p = 0.3
print("\n\nFMP para moneda truco (p_corona = 0.3): \n")
X_truco = binomial_fmp(n, p)
for x in range(len(X_ideal)):
    print("P( X =", x, ") = f(", x, ") = ", X_truco[x])
```

```python
# MONEDA IDEAL

_x = np.arange(n+1)

fig = plt.figure()
host = fig.add_subplot()
par = host.twiny()


host.bar(_x, X_ideal, width=0.04)
par.scatter(_x, X_ideal)

par.axes.get_xaxis().set_ticks([])
host.locator_params(axis="x", nbins=n+1)

host.set_xlabel("x")
host.set_ylabel("f(x)")
host.set_title("FMP para la moneda ideal")

plt.show()
```

```python
# MONEDA TRUCO

_x = np.arange(n+1)

fig = plt.figure()
host = fig.add_subplot()
par = host.twiny()


host.bar(_x, X_truco, width=0.04)
par.scatter(_x, X_truco)

par.axes.get_xaxis().set_ticks([])
host.locator_params(axis="x", nbins=n+1)

host.set_xlabel("x")
host.set_ylabel("f(x)")
host.set_title("FMP para la moneda truco")

plt.show()
```
