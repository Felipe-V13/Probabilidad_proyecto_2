### Probabilidad para el número de veces que se debe tirar la moneda antes de obtener la primera corona

```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
    
def fmp_geometrica(p, x):
    res = p*((1-p)**(x-1))
    return res

def media_geometrica(p):
    res = 1/p
    return res

def varianza_geometrica(p):
    res = (1-p)/p**2
    return res
    
X_geo_ideal = []
X_geo_truco = []
lanzamientos = 10

#MONEDA IDEAL
p = 0.5
#Recordar que no existe X = 0
for x in range(1, lanzamientos+1):
    X_geo_ideal.append(fmp_geometrica(p, x))
#Prints de moneda ideal    
cont=1
print("\nFPM moneda ideal distribución geométrica")
for x in X_geo_ideal:
    print("P( X =", cont, ") = f(", cont, ") =", x)
    cont = cont+1
print("Media: ", media_geometrica(p))
print("Varianza: ", varianza_geometrica(p))  
print("Desviación estándar: ", sqrt(varianza_geometrica(p)))    

#MONEDA TRUCO
p = 0.3
#Recordar que no existe X = 0
for x in range(1, lanzamientos+1):
    X_geo_truco.append(fmp_geometrica(p, x))    
#Prints de moneda truco
cont = 1
print("\nFPM moneda truco distribución geométrica")
for x in X_geo_truco:
    print("P( X =", cont, ") = f(", cont, ") =", x)
    cont = cont+1
print("Media: ", media_geometrica(p))
print("Varianza: ", varianza_geometrica(p))  
print("Desviación estándar: ", sqrt(varianza_geometrica(p)))
```

```python
# Graficamos para moneda ideal
ejex = np.arange(1, lanzamientos+1)
fig = plt.figure()
host = fig.add_subplot()
par = host.twiny()

host.bar(ejex, X_geo_ideal, width=0.04)   #puntos de gráfico
par.scatter(ejex, X_geo_ideal)            #barras de gráfico
par.axes.get_xaxis().set_ticks([])
host.locator_params(axis="x", nbins=lanzamientos)
host.set_title("fmp distribución geométrica para moneda ideal")
host.set_xlabel("x (lanzamiento)")
host.set_ylabel("f(x)")
```

```python
# Graficamos para moneda truco
ejex = np.arange(1, lanzamientos+1)
fig = plt.figure()
host = fig.add_subplot()
par = host.twiny()

host.bar(ejex, X_geo_truco, width=0.04)
par.scatter(ejex, X_geo_truco)
par.axes.get_xaxis().set_ticks([])
host.locator_params(axis="x", nbins=lanzamientos)
host.set_title("fmp distribución geométrica para moneda truco")
host.set_xlabel("x (lanzamiento)")
host.set_ylabel("f(x)")

plt.show()
```

### Probabilidad del número de veces que salga la cara de corona en las 10 tiradas del experimento

```python

def factorial(n):
    fact = 1
    for num in range(2, n + 1):
        fact *= num
    return fact

def binomial_fmp(n, p):
    X = []

    for x in range(n + 1):
        Comb_n_x = factorial(n)/(factorial(x) * factorial(n - x))
        X.append(Comb_n_x * (p**x) * ( (1-p)**(n-x) ))

    return X

def media_binomial(n, p):
    return n*p

def varianza_binomial(n, p):
    return n*p*(1 - p)

n = 10
p = 0.5 # moneda ideal

print("\nFMP para moneda ideal (p_corona = 0.5): \n")
X_ideal = binomial_fmp(n, p)
for x in range(len(X_ideal)):
    print("P( X =", x, ") = f(", x, ") = ", X_ideal[x])
print("Media: ", media_binomial(n,p))
print("Varianza: ", varianza_binomial(n,p))  
print("Desviación estándar: ", sqrt(varianza_binomial(n,p)))


p = 0.3 # moneda truco
print("\n\nFMP para moneda truco (p_corona = 0.3): \n")
X_truco = binomial_fmp(n, p)
for x in range(len(X_ideal)):
    print("P( X =", x, ") = f(", x, ") = ", X_truco[x])
print("Media: ", media_binomial(n,p))
print("Varianza: ", varianza_binomial(n,p))  
print("Desviación estándar: ", sqrt(varianza_binomial(n,p)))
```

```python
# MONEDA IDEAL

_x = np.arange(n+1)

fig = plt.figure()
host = fig.add_subplot()
par = host.twiny()


host.bar(_x, X_ideal, width=0.04)
par.scatter(_x, X_ideal)

par.axes.get_xaxis().set_ticks([])
host.locator_params(axis="x", nbins=n+1)

host.set_xlabel("x")
host.set_ylabel("f(x)")
host.set_title("FMP binomial para la moneda ideal")

plt.show()
```
